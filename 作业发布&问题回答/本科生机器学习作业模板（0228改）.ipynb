设有来自$A,B,C,D$4个地区的四名厨师制作名菜京城水煮鱼，想比较它们的品质是否相同。经四位美食评委评分结果如表4.21所示，试测验4个地区制作的水煮鱼这道菜品质有无区别。


## Assumption
【如果解题时有相关假设条件，需要一一列出】

由于不同评委在口味和美学欣赏上存在差异，因此适合用Friedman检验方法比较。
假设检验问题为：
$$H_0:4个地区的京城水煮鱼品质相同$$
$$H_1:4个地区的京城水煮鱼品质不同$$

```python
#示例代码，《非参数统计》例4.11
import numpy as np
import pandas as pd
from scipy.stats import rankdata
from scipy.stats import chi2
data=np.array([85,82,83,79,87,75,86,82,90,81,80,76,80,75,81,75]).reshape(4,4)
b=data.shape[0]
k=data.shape[1]
rankdat=np.array([list(rankdata(data[i])) for i in range(b)])#求秩矩阵

#求所有的Tao值
tao=[list(pd.value_counts(rankdat[i])) for i in range(b)]
taolist=[i for a in tao for i in a]

#计算Freidman检验统计量及p值
sumtao=sum([i**3-i for i in taolist])
sumR=sum([sum(rankdat[:,i])**2 for i in range(k)])
Q=12*sumR/(b*k*(k+1))-3*b*(k+1)
Qc=Q/(1-sumtao/(b*k*(k**2-1)))
pvalue=1-chi2.cdf(Qc,df=k-1)

print('Friedman chi-square is ',round(Qc,4))
print('p-value is',round(pvalue,4))
```
